```python
brand_counts = autos['brand'].value_counts(normalize=True)
top_brands = brand_counts[brand_counts > .05].index
print(top_brands)
```

These six brands make the top brands and  these 6 brands by number of listing each have more than 5% of the total number of listings. Together they represent 65% of the listings and are selected to be analyzed.

```python
brands_prices={}
for brands in top_brands:
    top_six=autos[autos['brand']==brands]
    mean_brands=top_six['price'].mean()
    brands_prices[brands]=int(mean_brands)
print(brands_prices)
```

```python
brand_mileage = {}

for brand in top_brands:
    group = autos[autos.brand == brand]
    brand_mileage[brand] = int(group.odometer_km.mean())
print(brand_mileage)
```

we aggregated across brands to understand mean price. We observed that in the top 6 brands, there's a distinct price gap.

* Audi, BMW and Mercedes Benz are more expensive
* Ford and Opel are less expensive
* Volkswagen is in between

```python
#Converting the dictoniries into series:
mean_mileage = pd.Series(brand_mileage).sort_values(ascending=False)
mean_price = pd.Series(brands_prices).sort_values(ascending=False)
```

```python
#creating a new data frame and assigning series to the new data frame:
brand_df=pd.DataFrame(mean_mileage,columns=['mean_mileage'])
```

```python
brand_df['mean_price'] = mean_price
```

```python
translation={
    'bus':'bus',
    'limousine':'limousine',
    'kleinwagen':'supermini',
    'kombi':'station_wagon',
    'coupe':'coupe',
    'suv':'suv',
    'cabrio':'cabrio',
    'andere' :'other',
    'manuell':'manual',
    'automatik':'automatic',
    'lpg':'lpg',
    'benzin':'petrol',
    'diesel':'diesel',
    'cng':'cng',
    'hybrid':'hybrid',
    'elektro':'electro',
    'nein':'no',
    'ja':'yes'
}
for each in ['vehicletype','gearbox','fueltype','unrepaired_damage']:
    autos[each] = autos[each].map(translation)
```
